```python
import os
import pickle

input_path = "../statistics/musdb18/agorski_sound_demixing"

for directory in os.listdir(input_path):
    path = os.path.join(input_path, directory, "statistics.pkl")
    if not os.path.exists(path):
        continue
    with open(path, "rb") as file:
        data = pickle.load(file)
    print(directory)
    print(max([list(x['median_sdr_dict'].values())[0] for x in data['test']]))
```
